```python
import pandas
import matplotlib.pyplot as plt
platy = pandas.read_csv("platy_2021_02.csv")
platy
```

```python
import matplotlib.pyplot as plt

platy['plat'].plot.hist(bins=10)
plt.xlabel('Plat')
plt.ylabel('Zaměstnanci')
plt.title('Rozdělení platů')
plt.show()
```
