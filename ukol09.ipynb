```python
import pandas as pd
teploty = pd.read_csv("temperature.csv")
teploty
```

```python
teploty.set_index("City", inplace = True)
teploty.head()
teploty.loc ["Prague"]
```

```python
teploty[teploty["AvgTemperature"] > 80]
```

```python
#Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně 
#bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

evropa_nejvyssi_teplota = teploty[(teploty["AvgTemperature"] > 60) & (teploty["Region"] == "Europe")]
print(evropa_nejvyssi_teplota["AvgTemperature"])
```

```python
#Dotaz na extrémní hodnoty, 
# tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než -20 stupňů.

extremni_teploty = teploty[(teploty["AvgTemperature"] > 80) | (teploty["AvgTemperature"] < -20)]
print(extremni_teploty["AvgTemperature"])
```
